# SphericalStreamerCow
Toy model for C/O ratio evolution in protoplanetary disks with stochastic streamer infall.

Set parameters below, then click **▶ Run simulation**.

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import ipywidgets as widgets
from IPython.display import display
from PIL import Image, ImageEnhance
```

```python
T_MAX    = 2e6   # yr
DT       = 1e3   # yr
T0_RATE  = 1e5   # yr
M_S_MIN  = 1e-9  # M_sun
M_S_MAX  = 0.1   # M_sun
DUR_MIN  = 1e3   # yr
DUR_MAX  = 1e4   # yr
CO_FIXED = 0.4   # fixed C/O for disk initial and streamers


def build_rate(times, decay_fn):
    t = times / T_MAX
    if decay_fn == 'linear':
        return np.maximum(0.0, 1.0 - t)
    elif decay_fn == 'quadratic':
        return np.maximum(0.0, (1.0 - t) ** 2)
    else:  # exponential
        r = np.exp(-5.0 * t); r[-1] = 0.0; return r


def run_simulation(M0, COf, M_res_0, n0, decay_fn, seed):
    CO0 = CO_FIXED
    CO_s = CO_FIXED

    rng   = np.random.default_rng(seed)
    times = np.arange(0, T_MAX + DT, DT)
    N     = len(times)

    rate        = build_rate(times, decay_fn)
    steps_in_T0 = int(T0_RATE / DT)
    exp_in_T0   = rate[:steps_in_T0].sum()
    prob = np.clip(rate * (n0 / exp_in_T0), 0, 1) if exp_in_T0 > 0 else np.zeros(N)

    events = []  # (i_start, dur_steps, dm_per_yr, mass_requested)
    for i, p in enumerate(prob):
        if rng.random() < p:
            mass  = 10 ** rng.uniform(np.log10(M_S_MIN), np.log10(M_S_MAX))
            t_log = (np.log10(mass) - np.log10(M_S_MIN)) / (np.log10(M_S_MAX) - np.log10(M_S_MIN))
            dur_s = max(1, int(round(DUR_MIN * (DUR_MAX / DUR_MIN) ** t_log / DT)))
            events.append((i, dur_s, mass / (dur_s * DT), mass))

    Mdisk  = np.zeros(N);  Mdisk[0]  = M0
    COdisk = np.zeros(N);  COdisk[0] = CO0
    Mres   = np.zeros(N);  Mres[0]   = M_res_0
    dM_dt  = -M0 / T_MAX
    dCO_dt = (COf - CO0) / T_MAX

    for i in range(1, N):
        M   = max(0.0, Mdisk[i-1] + dM_dt * DT)
        CO  = COdisk[i-1] + dCO_dt * DT
        res = Mres[i-1]
        for (i_s, dur_s, dm_yr, _) in events:
            if i_s < i <= i_s + dur_s and res > 0:
                dm  = min(dm_yr * DT, res)
                CO  = (M * CO + dm * CO_s) / (M + dm)
                M  += dm
                res = max(0.0, res - dm)
        Mdisk[i]  = M
        COdisk[i] = CO
        Mres[i]   = res

    event_meta = [(i_s, i_s + dur_s, mass) for (i_s, dur_s, _, mass) in events]
    return times, Mdisk, COdisk, Mres, event_meta


def brightened(img, factor):
    return np.array(ImageEnhance.Brightness(img.convert('RGB')).enhance(factor))


_cow = Image.open('spherical_cow.gif').convert('RGBA')
CW, CH = _cow.size
print(f'Cow image: {CW}×{CH} px')
```

```python
_W = '420px'
_S = {'description_width': '175px'}

w_M0    = widgets.FloatSlider(value=0.1, min=0.01, max=0.5,  step=0.01,
    description='M_disk,0 [M☉]',          style=_S, layout=widgets.Layout(width=_W))
w_COf   = widgets.FloatSlider(value=1.5, min=0.5,  max=3.0,  step=0.05,
    description='C/O disk, final',         style=_S, layout=widgets.Layout(width=_W))
w_Mres  = widgets.FloatSlider(value=0.1, min=0.0,  max=0.5,  step=0.005,
    description='M_reservoir [M☉]',        style=_S, layout=widgets.Layout(width=_W))
w_n0    = widgets.IntSlider(  value=5,   min=0,    max=30,   step=1,
    description='N streamers (100 kyr)',   style=_S, layout=widgets.Layout(width=_W))
w_decay = widgets.Dropdown(
    options=[('Linear','linear'),('Quadratic','quadratic'),('Exponential','exponential')],
    value='quadratic', description='Rate decay', style=_S)
w_seed  = widgets.IntText(value=42, description='Random seed',
    style=_S, layout=widgets.Layout(width='280px'))
btn = widgets.Button(description='▶  Run simulation', button_style='success',
    layout=widgets.Layout(width='200px', height='36px'))
out = widgets.Output()

display(
    widgets.HTML('<b>── Disk ──────────────────────────</b>'),
    w_M0, w_COf,
    widgets.HTML('<b>── Streamers ────────────────────</b>'),
    w_Mres, w_n0, w_decay,
    widgets.HTML('<b>── Simulation ───────────────────</b>'),
    w_seed, btn, out,
)
```

```python
def fmt_mass(m):
    if m >= 0.01:
        return f'{m:.3f} M☉'
    elif m > 1e-12:
        return f'{m:.2e} M☉'
    else:
        return '0 M☉  (depleted)'


def on_run(_btn):
    out.clear_output(wait=True)
    with out:
        # ── Run simulation ───────────────────────────────────
        times, Mdisk, COdisk, Mres, event_meta = run_simulation(
            w_M0.value, w_COf.value, w_Mres.value,
            w_n0.value, w_decay.value, w_seed.value)

        CO0_eff   = CO_FIXED
        COf_eff   = w_COf.value
        times_kyr = times / 1e3
        N = len(times)
        print(f'{len(event_meta)} streamer event(s)  |  '
              f'reservoir at end: {fmt_mass(Mres[-1])}')

        # Pre-assign stable visual properties per event
        rng_vis = np.random.default_rng(w_seed.value + 1)
        evt_vis = []
        for (_, _, mass) in event_meta:
            cx = rng_vis.uniform(0.1 * CW, 0.9 * CW)
            cy = rng_vis.uniform(0.1 * CH, 0.9 * CH)
            r_norm = (np.log10(mass) - np.log10(M_S_MIN)) / (np.log10(M_S_MAX) - np.log10(M_S_MIN))
            radius = 10 + r_norm * 50
            color  = rng_vis.random(3).tolist()
            evt_vis.append((cx, cy, radius, color))

        # ── Build figure ─────────────────────────────────────
        plt.close('all')
        fig, axes = plt.subplots(1, 3, figsize=(13, 4.5))
        ax_m, ax_co, ax_cow = axes
        fig.tight_layout(pad=3.0)

        # Faint full-history reference traces
        ax_m .plot(times_kyr, Mdisk,  color='steelblue', alpha=0.15, lw=1)
        ax_co.plot(times_kyr, COdisk, color='tomato',    alpha=0.15, lw=1)

        line_m,  = ax_m .plot([], [], color='steelblue', lw=2)
        dot_m,   = ax_m .plot([], [], 'o', color='steelblue', ms=6, zorder=5)
        line_co, = ax_co.plot([], [], color='tomato',    lw=2)
        dot_co,  = ax_co.plot([], [], 'o', color='tomato',    ms=6, zorder=5)

        ax_m.set(xlabel='Time [kyr]', ylabel='Disk mass [M☉]', title='Disk mass',
                 xlim=(0, times_kyr[-1]), ylim=(0, Mdisk.max() * 1.15))
        ax_co.set(xlabel='Time [kyr]', ylabel='C/O ratio', title='Disk C/O ratio',
                  xlim=(0, times_kyr[-1]),
                  ylim=(COdisk.min() * 0.85, COdisk.max() * 1.1))
        ax_co.axhline(CO_FIXED, color='gray', ls='--', lw=0.8, label=f'C/O = {CO_FIXED}')
        ax_co.legend(fontsize=7)

        ax_cow.set_title('Spherical Cow')
        ax_cow.axis('off')
        ax_cow.set_xlim(0, CW)
        ax_cow.set_ylim(CH, 0)
        cow_im = ax_cow.imshow(brightened(_cow, 1.0),
                               extent=[0, CW, CH, 0], zorder=0)

        # KEY FIX: use display(fig.canvas) instead of plt.show()
        # inside a widgets.Output() context with the widget backend
        display(fig.canvas)

        # ── Playback controls ────────────────────────────────
        ANIM_SECS   = 45
        interval_ms = max(20, int(ANIM_SECS * 1000 / N))
        play   = widgets.Play(value=0, min=0, max=N-1, step=1,
                              interval=interval_ms, description='▶')
        slider = widgets.IntSlider(value=0, min=0, max=N-1, step=1,
                                   description='',
                                   layout=widgets.Layout(width='450px'))
        info   = widgets.Label(value='', layout=widgets.Layout(width='340px'))
        widgets.jslink((play, 'value'), (slider, 'value'))

        active_patches = []

        def draw_frame(i):
            line_m .set_data(times_kyr[:i+1], Mdisk[:i+1])
            dot_m  .set_data([times_kyr[i]], [Mdisk[i]])
            line_co.set_data(times_kyr[:i+1], COdisk[:i+1])
            dot_co .set_data([times_kyr[i]], [COdisk[i]])

            bright = 1.0 - 0.8 * np.clip(
                (COdisk[i] - CO0_eff) / max(COf_eff - CO0_eff, 1e-9), 0, 1)
            cow_im.set_data(brightened(_cow, bright))

            for p in active_patches:
                p.remove()
            active_patches.clear()
            for k, (i_s, i_e, _mass) in enumerate(event_meta):
                if i_s < i <= i_e:
                    cx, cy, radius, color = evt_vis[k]
                    p = mpatches.Circle((cx, cy), radius,
                                        color=color, alpha=0.75, zorder=5)
                    ax_cow.add_patch(p)
                    active_patches.append(p)

            info.value = (f't = {times_kyr[i]:.0f} kyr   |   '
                          f'reservoir: {fmt_mass(Mres[i])}')
            fig.canvas.draw_idle()

        def on_slider(change):
            draw_frame(change['new'])

        slider.observe(on_slider, names='value')
        draw_frame(0)

        display(widgets.HBox([play, slider]), info)


btn.on_click(on_run)
```
